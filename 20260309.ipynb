```python
class FourCal:
    # 클래스에서 함수를 사용할 때는 반드시 self 매개변수 사용
    # self -> 실행 시점점의 함수를 포함하는 객체 자신을 가르키는 변수
    def __init__(self, first, second): # 변수 초기화
        self.first = first
        self.second = second
    
    def add(self): # 함수
        return self.first + self.second
    def sub(self):
        return self.first - self.second
    def mul(self):
        return self.first * self.second
    def div(self):
        return self.first / self.second

```

```python
cal1 = FourCal() # 객체 생성 => 프로그램에서 사용가능한 상태(메모리에 변수와 함수가 올라가는 상태)

cal1.setData(10,30)

print(cal1.add(),cal1.sub(),cal1.mul(),cal1.div())
```

### 클래스의 상속

```python
# 클래스의 상속 : 새로운 클래스를 생성
# 기존의 클래스의 맴버(변수, 메서드)들을 받아서 새로운 클래스 생성
# 오버라이딩 : 메서드를 수정(처리블록)해서 사용
# MoreFourlCal : 새로운 클래스 정의 -> pow 제곱 구해주는 함수, div 메서드를 수정
class MoreFourCal(FourCal):
    def pow(self):
        return self.first ** self.second
    
    #  오버라이딩: 함수의 이름과 매개변수는 그대로 하고 처리부붑ㄴ을 수정해서 정의
    def div(self):
        if self.second == 0 :
            return 0
        return self.first/self.second
```

```python
mfc1 = MoreFourCal(10,3)
print(mfc1.add())  # 13
print(mfc1.pow())  # 1000

mfc1.div()
```
