```python
# format 함수

print("I eat {0} apples".format(3))
f_apple = 3
print("I eat {0} apples".format(f_apple))
print("I eat {0} apples".format("four"))
f_number = 8
f_day = "five"
print("I ate {0} apples. so I was sick for {1} days".format(f_number, f_day))
print("I ate {f_number} apples. so I was sick for {f_day} days".format(f_number=10, f_day=3))
print("I ate {0} apples. so I was sick for {f_day} days".format(10, f_day=3))
print("{0:<10}hello".format("hi"))
print("hello{0:>10}".format("HI"))
print("hi{0:^10}bye".format("hi"))
print("{0:=^10}".format("hi"))
print("{0:!<10}".format("hi"))
f_ex = 3.141592
print("{0:0.4f}".format(f_ex))
print("{0:10.4f}".format(f_ex))
print("{{ and }}".format())

```

```python
# f 문자열 포매팅 

f_name = '최준희'
f_age = 28
print(f"나의 이름은 {f_name} 입니다. 나이는 {f_age} 입니다.")
f_d = {'name':'최준희', 'age':28}
print(f"나의 이름은 {f_d['name']} 입니다. 나이는 {f_d['age']} 입니다.")
print(f'{"hi":<10}')
print(f'{"hi":>10}')
print(f'{"hi":^10}')
print(f'{"hi":=^10}')
print(f'{"hi":!>10}')
print(f'{f_ex:0.4f}')
print(f'{f_ex:10.4f}')
```

```python
set_1 = set([1,2,3,4,5])
sset_3 = set([1,2,3,4,5,6])
set_4 = set([4,5,6,7,8,9])
print(set_3 & set_4) # {4, 5, 6}
print(set_3 | set_4) # {1, 2, 3, 4, 5, 6, 7, 8, 9}
print(set_3 - set_4) # {1, 2, 3}
print(set_4 - set_3)
```
