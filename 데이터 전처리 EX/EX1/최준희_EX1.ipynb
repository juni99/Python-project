### 문제 1: 결측치 처리

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("train.csv") # train.csv 파일을 DataFrame 형태로 해서 변수 저장

# Age 결측치 -> 평균값으로 대체
df["Age"] = df["Age"].fillna(df["Age"].mean())

# Embarked 결측치 -> 최빈값으로 대체
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# 결측치가 많은 Cabin 열 삭제
df = df.drop(columns=["Cabin"])

# 결측치 확인
print(df.isnull().sum())
```
