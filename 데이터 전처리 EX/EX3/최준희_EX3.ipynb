### 데이터 전처리 EX-3

```python
import pandas as pd

df = pd.read_csv("diamonds.csv")

# Q1, Q3 계산
Q1 = df["carat"].quantile(0.25) # Q1(25% 분위수)
Q3 = df["carat"].quantile(0.75) # Q3(75% 분위수)

# IQR 계산
IQR = Q3 - Q1
print(Q1, Q3, IQR)
print('='*40)

# 이상치 경계 계산
lower_bound = Q1 - 1.5 * IQR # 하한 값
upper_bound = Q3 + 1.5 * IQR # 상한 값

print("Lower Bound:", lower_bound) 
print("Upper Bound:", upper_bound)
print('='*80)

# 이상치 제거
df_clean = df[(df["carat"] >= lower_bound) & (df["carat"] <= upper_bound)]
print(df_clean)
print('='*80)

# 데이터 개수 출력
print("이상치 제거 후 데이터 개수:", len(df_clean))
```
