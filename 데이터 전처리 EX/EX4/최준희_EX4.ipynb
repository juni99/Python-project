### 데이터 전처리 EX-4

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("adult.csv")

# Label Encoding 할 컬럼 리스트
columns = [
    "workclass",
    "education",
    "marital-status",
    "occupation",
    "relationship",
    "race",
    "gender",
    "native-country"
]

# LabelEncoder 객체 생성
le = LabelEncoder()

# 각 컬럼에 Label Encoding 적용
for col in columns:
    df[col] = le.fit_transform(df[col])

df.head()
```
