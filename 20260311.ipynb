```python
import requests
# 요청 URL
url = 'https://jsonplaceholder.typicode.com/posts'
# 요청 메서드 : get 구현한 기능 => get()
response = requests.get(url)

if response.status_code == 200 :
    print('정상적인 응답 완료')
    print(response.text[:200])  # JSON ( TEXT )
else : # 404, 5XX
    print('정상적인 수신이 되지 않았음')
    print(f'오류발생 : {response.status_code} ')
```

```python
from bs4 import BeautifulSoup
```

```python
html_doc = """
<html>
    <head>
        <title>Example Page</title>
    </head>
    <body>
        <h1 class="description">Welcome to Web Scraping</h1>
        <p class="description title big">This is an example paragraph.</p>
        <a href="https://example.com1">Visit Example1</a>
        <a href="https://example.com2">Visit Example2</a>
        <a href="https://example.com3">Visit Example3</a>
        <a href="https://example.com4">Visit Example4</a>
    </body>
</html>
"""
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
print(soup.title)
print(soup.title.string)
print(soup.h1)
print(soup.h1.string)
print(soup.h1['class'])
print(soup.a)
print(soup.a['href'])
print(soup.p)
print(soup.p['class'])
```

```python
# find(tag), find_all(tag)
a_tag = soup.find('a')
print(a_tag)
a_tags = soup.find_all('a')
print(a_tags)
```

```python
links = []  # link url list
for a_tag in a_tags :
    print(a_tag.string, a_tag['href'])
    links.append(a_tag['href'])

print(links)
```

SQLite : 관계형데이터베이스

```python
import sqlite3
```

```python
# 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')
conn.close()
```

```python
# 관계형데이터베이스의 특징
# 테이블 형식으로 데이터 관리
# 저장할 공간 생성 => 테이블을 생성
sql_create = '''
create table user (
    id integer primary key autoincrement,
    name text not null,
    age intger,
    city text
)
'''
# 1. 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')

# 2. sql을 실행 시키기 위한 객체가 필요
cursor = conn.cursor()

#. 3. sql 실행
cursor.execute(sql_create)
print('테이블 생성!')

# 4. 연결 종료
conn.close()
```

```python
# 데이터 입력 : 행단위 입력
# id, name, age, city
sql_insert ='''
insert into user (name, age, city) values (?,?,?)
'''
# 1. 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')

# 2. sql을 실행 시키기 위한 객체가 필요
cursor = conn.cursor()

data = [
    ('hot1', 10, 'pusan1'),
    ('hot2', 25, 'pusan2'),
    ('hot3', 32, 'pusan3')
]

#. 3. sql 실행
# insert into user (name, age, city) values (?,?,?)
#cursor.execute(sql_insert, ('hot', 30, 'pusan'))
cursor.executemany(sql_insert, data)

# 작업완료 처리 : 입력, 수정, 삭제 저리시에 작업 완료 또는 롤백을 지정해주어야 한다!!!!
# commit: 완료
conn.commit() # 물리적인 데이터 저장

# 4. 연결 종료
conn.close()
```

```python
# 검색(질의) : 행단위 검색
sql_select = 'select * from user'
sql_select_where = 'select * from user where id>2 and age>20'
sql_select_where_col = 'select name, city from user where id>2 and age>20'

# 1. 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')

# 2. sql을 실행 시키기 위한 객체가 필요
cursor = conn.cursor()

#. 3. sql 실행
# select 결과 => 테이블 , 
# fetchall() : 전체 행 반환,  fetchone() : 행 하나 반환
cursor.execute(sql_select)
rows = cursor.fetchall()
#print(rows)

for row in rows :
    print(row)


# 4. 연결 종료
conn.close()
```

```python
# 데이터 수정
# 행을 찾고 -> 해당 행의 컬럼값을 변경
sql_update = 'update user set age=?, name=? where id=?'

# 1. 데이터 베이스 연결
conn = sqlite3.connect('first.db')
print('데이터베이스 연결 성공!')

# 2. sql을 실행 시키기 위한 객체가 필요
cursor = conn.cursor()

# 3. sql 실행
# update user set age=?, name=? where id=?
cursor.execute(sql_update, (40, 'king', 1))

conn.commit()
print('수정완료!')
conn.close()
```

```python
# 데이터 삭제 : 행단위 삭제
sql_delete = 'delete from user where id=?'
conn = sqlite3.connect('first.db')
cursor = conn.cursor()
cursor.execute(sql_delete, (1,))
conn.commit()
print('삭제완료!')
conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fulldata_03_13_02_P_영화상영관.csv', encoding='cp949', low_memory=False)


data_1 = df[['번호', '영업상태명', '사업장명', '소재지전체주소', '도로명전체주소']]

data_1 = data_1.dropna(subset=['소재지전체주소'])

data_2 = data_1[ data_1['소재지전체주소'].str.contains('서울특별시') ]

data_3 = data_2[ data_2['사업장명'].str.contains('메가박스')]

data_4 = data_3[data_3['영업상태명'] != '폐업']

data_4.loc[:, '시군구'] = data_4['소재지전체주소'].str.split().str[1]

data_5 = data_4.groupby('시군구').size().reset_index(name='사업장수')


plt.rcParams['font.family'] = 'Batang'
plt.bar(
    data_5['시군구'],
    data_5['사업장수']
)
plt.xticks(rotation=45)
plt.title('')
```
