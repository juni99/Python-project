```python
num = int(input('숫자를 입력하세여'))
if num%2 == 0 : # 입력받은 num 변수가 2로 나눈 나머지가 2라면
    print('짝수')
else :
    print('홀수')
```

```python
num = int(input('숫자를 입력하세여'))
if num > 0 :
    print('양수')
elif num < 0 :
    print('음수')
else :
    print('0')
```

```python
age = int(input('나이를 입력하세여'))
if age < 15 or age >= 65 :
    print('무료 예방 접종이 가능.')
else :
    print('무료 예방 접종이 불가능.')
```

```python
year = 2026
birth = int(input('태어난 연도를 입력하세여(4자리)'))
age = year - birth

if age >= 20 and age%2 == 0 :
    print('무료 건강 검진 가능')
    if age >= 40 :
        print('무료 암 검사 가능')
else :
    print('무료 건강 검진 불가능')
```
