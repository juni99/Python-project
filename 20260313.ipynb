```python
import pandas as pd
```

```python
# 고객 정보 데이터프레임
customers = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})
# 주문 정보 데이터프레임
orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 4],
    'product': ['Laptop', 'Tablet', 'Smartphone']
})
```

```python
print(customers)
print(orders)
```

```python
# merge : 열단위 데이터 병합 => 옆으로 열을 붙여서 새로운 표를 만든다!
# 공통된 customer_id로 데이터 병합 : Inner Join
result1 = pd.merge(customers, orders, on='customer_id', how='inner')
result1
```

```python
result2 = pd.merge(customers, orders, on='customer_id', how='right')
result2
```

```python
# 그룹핑
# 샘플 데이터
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 55000, 70000, 65000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
grouped_df = df.groupby('Department')['Salary'].mean()
grouped_df
```

```python
pivot_table = pd.pivot_table(
    df,
    values='Salary',
    index='Department',
    aggfunc=['mean', 'sum']
)
pivot_table
```

```python
# 새로운 df
data_1 = df[['번호', '영업상태명', '사업장명', '소재지전체주소', '도로명전체주소']]
data_1
```

```python
# 소재지 기준으로 값 NaN인 값을 삭제
data_1 = data_1.dropna(subset=['소재지전체주소'])
data_1.info()
```

```python
# 소재지 기준으로 '서울특별시' 를 포함하는 행만 남기는 처리
data_2 = data_1[ data_1['소재지전체주소'].str.contains('서울특별시') ]
data_2.info()
```

```python
data_2.head()
```

```python
# 사업장 명 => '스타벅스'
data_3 = data_2[ data_2['사업장명'].str.contains('스타벅스')]
data_3.info()
```

```python
data_3
```

```python
data_4 = data_3[data_3['영업상태명'] != '폐업']
data_4.info()
```

```python
data_4
```

```python
'서울특별시 마포구 대흥동 111-1'.split()[1]
```

```python
# 열을 추가 : 시군구
# 구별로 스타벅스 매장이 몇개가 있는지 
data_4.loc[:, '시군구'] = data_4['소재지전체주소'].str.split().str[1]
data_4
```

```python
data_5 = data_4.groupby('시군구').size().reset_index(name='사업장수')
data_5
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'Batang'
plt.bar(
    data_5['시군구'],
    data_5['사업장수']
)
plt.xticks(rotation=45)
plt.title('')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('fulldata_03_13_02_P_영화상영관.csv', encoding='cp949', low_memory=False)


data_1 = df[['번호', '영업상태명', '사업장명', '소재지전체주소', '도로명전체주소']]

data_1 = data_1.dropna(subset=['소재지전체주소'])

data_2 = data_1[ data_1['소재지전체주소'].str.contains('서울특별시') ]

data_3 = data_2[ data_2['사업장명'].str.contains('메가박스')]

data_4 = data_3[data_3['영업상태명'] != '폐업']

data_4.loc[:, '시군구'] = data_4['소재지전체주소'].str.split().str[1]

data_5 = data_4.groupby('시군구').size().reset_index(name='사업장수')


plt.rcParams['font.family'] = 'Batang'
plt.bar(
    data_5['시군구'],
    data_5['사업장수']
)
plt.xticks(rotation=45)
plt.title('')
```
