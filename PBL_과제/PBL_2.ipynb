```python
# 메뉴와 가격표
menu = {"아메리카노": 1500,"라떼": 4500,"녹차": 4500,"케이크": 15000}

order = [] # 주문 저장용 빈 리스트

while True: # 무한 반복
    call = input("메뉴를 입력하세요(추가할 메뉴가 없다면 끝 을 입력하세요): ")

    if call == "끝":
        break

    if call in menu:
        order.append(call)
        print(call, "추가되었습니다.")
    else:
        print("없는 메뉴입니다.")

total = 0 # 주문 합계 가격
for a in order:
    total += menu[a]

count = len(order)

if total >= 20000: # 총 가격이 20000 이상일 시 10% 할인
    discount = total * 0.1
else:
    discount = 0

final_price = total - discount

menu_type = set(order)

for m in menu_type: # 메뉴별 주문 횟수
    print(m, "주문 횟수:", order.count(m))

print("주문 목록:", order)
print("총 주문 수:", count)
print("메뉴 종류:", menu_type)
print("합계:", total, "원")
print("할인:", discount, "원")
print("최종 결제:", final_price, "원")
```
