```python
apple_name = "사과"
apple_price = 1600
apple_num = 5

water_name = "물"
water_price = 500
water_num = 10

cookie_name = "쿠키"
cookie_price = 3000
cookie_num = 3

apple = {'name': apple_name, 'price':apple_price, 'num': apple_num}
water = {'name': water_name, 'price':water_price, 'num': water_num}
cookie = {'name': cookie_name, 'price':cookie_price, 'num': cookie_num}

cart = [apple, water, cookie]

apple_total = apple['price'] * apple['num'] # 사과 5개의 가격
water_total = water['price'] * water['num'] # 물 10개의 가격
cookie_total = cookie['price'] * cookie['num'] # 쿠키 3개의 가격

delivery_fee = 3000 # 배송비

total_num = apple['num'] + water['num'] + cookie['num'] # 물품 총 수량
subtotal = apple_total + water_total + cookie_total # 물품 총 가격
final_total = subtotal + delivery_fee # 배송비 포함한 총 가격
avg_price = subtotal / total_num # 물품 한 개당 평균 가격
has_discount = final_total >= 10000 or final_total < 10000 # 총 구매 가격이 10000 이상일 경우 할인 대상
summary = f"장바구니: {apple_name} {apple_num}개, {water_name} {water_num}개, {cookie_name} {cookie_num}개 | 상품합계: {subtotal} | 배송비: {delivery_fee} | 최종: {final_total} | 할인대상: {has_discount}"

print(summary)
```
