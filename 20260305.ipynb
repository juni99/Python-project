```python
# while

cnt = 1

while cnt < 6 :
    print(cnt)
    cnt = cnt + 1
```

```python
# continue 반복문 내에서 이 키워드를 만나면 아래 코드들을 실행하지 않고 블록 종료
cnt = 0
while cnt < 10 :
    cnt += 1
    if cnt%2 == 0 :
        continue
    print(cnt)
```

```python
# break : 반복문 내에서 이 키워드를 만나면 반복문이 종료됨.
cnt = 0
while True :
    cnt += 1
    print(cnt)
    if cnt > 4:
        break
```

```python
scores = [60, 50, 70, 100, 49]
num = 0 # 학생 일련 번호
total = 0
for i in scores :
    num += 1
    msg = 'PASS' if i >= 60 else "FAIL" # 점수가 60점 이상이면 msg 변수에 "PASS"를 저장.
    print(num, i, msg)
    total = total + i
print("총점 :",total)
print("평균 :",total/len(scores))
```

```python
def add_all_sum(*args):
    # print(args)
    # print(type(args))
    sum = 0
    for i in args:
        # print(i, end=', ')
        # print(type(i))
        sum += i
    return sum

add_all_sum(1,2,3,4,5,6,7,8,9,0)
```

```python
def proc_type(proc_type, *args):
    print(proc_type, args)
    if proc_type =='+':
        result = 0
        for num in args:
            result += num
        return result
    elif proc_type =='*':
        result = 1
        for num in args:
            result *= num
        return result
    else :
        print('연산이 가능한 기호를 입력하세여(+ 또는 *)')

proc_type('+', 1,2,3,4,5)
```
