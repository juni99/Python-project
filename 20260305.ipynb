```python
# while

cnt = 1

while cnt < 6 :
    print(cnt)
    cnt = cnt + 1
```

```python
# continue 반복문 내에서 이 키워드를 만나면 아래 코드들을 실행하지 않고 블록 종료
cnt = 0
while cnt < 10 :
    cnt += 1
    if cnt%2 == 0 :
        continue
    print(cnt)
```

```python
# break : 반복문 내에서 이 키워드를 만나면 반복문이 종료됨.
cnt = 0
while True :
    cnt += 1
    print(cnt)
    if cnt > 4:
        break
```

```python
# for 시퀀스 데이터를 반복처리할 대 사용

list_a = [1, 2, 3, 4, 5, 100, 200, 300, 1000]
print(len(list_a))
for i in list_a :
    print(i, end = ', ')
```

```python
scores = [60, 50, 70, 100, 49]
num = 0 # 학생 일련 번호
total = 0
for i in scores :
    num += 1
    msg = 'PASS' if i >= 60 else "FAIL" # 점수가 60점 이상이면 msg 변수에 "PASS"를 저장.
    print(num, i, msg)
    total = total + i
print("총점 :",total)
print("평균 :",total/len(scores))
```

```python
list_c = [1,2,3,4]
list_result = []

for num in list_c:
    if num%2==1:
        print(num)
        list_result.append(num*3)

print(list_result)
```
