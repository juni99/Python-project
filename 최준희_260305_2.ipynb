```python
# 1 부터 100 까지의 합
num = 0

for i in range(1,101):
    num += i

print(num)
```

```python
# 1부터 10까지의 곱셈
num = 1

for i in range(1,11):
    num *= i

print(num)
```

```python
for dan in range(2, 10, 2):
    print(f'{dan} 단')
    for num in range(1, dan+1) :
        print(f'{dan} X {num}={dan*num}')
```

```python
num = 0
total = 0

while True :
    num += 1
    if num % 2 == 0 or num % 3 == 0:
        total += num
    if total > 1000 :
            break

print(total)
```
